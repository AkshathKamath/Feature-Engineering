#### For multiple categories in a feature





```python
merc_df=pd.read_csv('mercedes.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
merc_df.head()
```

```python
for col in merc_df.columns: # No. of catgeories per feature
    print(len(merc_df[col].unique()))
```

```python
## For X0
merc_df['X0'].value_counts().head(10) # 10 most freq. categories (KDD cup orange challenge strategy)
```

```python
list_10=merc_df['X0'].value_counts().head(10).index
list_10=list(list_10)
list_10
```

```python
for category in list_10:
    merc_df[category]=np.where(merc_df['X0']==category,1,0)
list_10.append('X0')
merc_df[list_10].head(10)
```
